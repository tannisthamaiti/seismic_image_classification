```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cv2
import os
```

```python
#reading data and labels
seismic_test = np.load('test_once/test1_seismic.npy')
labels_test = np.load('test_once/test1_labels.npy')
```

```python
data_file = open('test.txt', 'r')
for line in data_file:
    print(line)
    patch_name = line
    direction, number = patch_name.split(sep='_')
    number = int(number)
    print(number)
    if direction == 'i':
        im = seismic_test[int(number), :, :]
        lbl =labels_test[int(number), :, :]
    elif direction == 'x':    
        im = seismic_test[:, int(number), :]
        lbl = labels_test[:, int(number), :]
    my_string = 'testX/seismic/'+ str(direction)+ '_' + str(number)
    my_string1 = 'testX/labels/' + str(direction)+ '_' + str(number) 
    plt.imsave('{}.png'.format(my_string), im)
    plt.imsave('{}.png'.format(my_string1), lbl)      
```

```python
#code to rotate and store generated images and their labels
input_folder = "testX/seismic/"
output_folder = "testX/seismic/"

for i in os.listdir(input_folder):
    img = cv2.imread(input_folder + i)
    img_rotate_90_counterclockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    cv2.imwrite(output_folder + i, img_rotate_90_counterclockwise)
```

```python
#code to rotate and store generated images and their labels
input_folder = "testX/labels/"
output_folder = "testX/labels/"

for i in os.listdir(input_folder):
    img = cv2.imread(input_folder + i)
    img_rotate_90_counterclockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    cv2.imwrite(output_folder + i, img_rotate_90_counterclockwise)
```
